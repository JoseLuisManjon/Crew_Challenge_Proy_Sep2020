Me descargo el json con la lista de los restaurantes que tienen datos de los platos.

```python
rest_26 = pd.read_json("json_rest.json")
rest_26.head()
```

```python
lista_26 = list(rest_26.restaurants.unique())
```

```python
# Compruebo los que han
df[df['Restaurant'].isin(lista_26)]
```

```python
# Con esto compruebo si los 26 restaurantes de los que tengo el menú están en la lista o no
mas_restaurantes = df[~df['Restaurant'].isin(lista_26)]
mas_restaurantes
```

```python
mas_restaurantes.reset_index(drop=True, inplace=True) # Sustituyo indice para que aparezcan valores contiguos y crecientes
```

```python
mas_restaurantes.head(3)
```

```python
# Convierto el df a json y lo guardo
json_mas_rest = mas_restaurantes.to_json(orient='records')
parsed = json.loads(json_mas_rest)
with open('json_mas_rest.json', 'w+') as outfile:
    json.dump(parsed, outfile, indent=4)
```
